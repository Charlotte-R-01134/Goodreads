```python
# get data from json file
import json

def get_data():
    with open('../webscraper/data/books.json', 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data

data = get_data()
```

```python
import pandas as pd

df = pd.DataFrame(data)
df = df.transpose()
df.head()
```

```python
import nltk
nltk.download('punkt')
from nltk.probability import FreqDist
from collections import defaultdict
```

```python
def get_most_liked_genres(ratings=[4,5]):
    weights = {rating: 1+i for i, rating in enumerate(ratings)}
    genres = defaultdict(int)
    for i in range(len(df)):
        if df['rating'][i] in ratings:
            for genre in df['genres'][i]:
                if genre != "general":
                    genres[genre] += weights[df['rating'][i]]
    return FreqDist(genres)

# most = get_most_liked_genres()
# for genre, count in most.most_common(20):
#     print(genre, count)
```
